```python
state_respondents = df['US_State'].value_counts()
state_uses_SQL = state_grp['LanguageHaveWorkedWith'].apply(lambda x: x.str.contains('SQL').sum())
state_data_career = state_grp['DevType'].apply(lambda x: x.str.contains('Data').sum())
```

```python
sql_df = pd.concat([state_respondents, state_uses_SQL, state_data_career], axis='columns', sort=True)
sql_df.rename(columns={'US_State': "NumRespondents", 'LanguageHaveWorkedWith': 'LanguageHaveWorkedWith:SQL', 'DevType':'DataField'}, inplace=True)
```

```python
sql_df.head(10)
```

```python
sql_df['PercentageKnowSQL'] = (sql_df['LanguageHaveWorkedWith:SQL']/sql_df['NumRespondents']) *100
sql_df['PercentageInDataField'] = (sql_df['DataField']/sql_df['NumRespondents']) *100
```
