### 1. What percentage of all licensed sets ever released were Star Wars themed?

```python
# Join both tables by name

merged = df.merge(theme, left_on='parent_theme', right_on='name')

# Drop name_x since both tables have column name "names"
merged.drop(columns='name_y', inplace=True)

```

```python
merged.head()
```

```python
# Missing values

percent_missing = merged.isnull().sum() * 100 / len(df)
missing_value_df = pd.DataFrame({'percent_missing': percent_missing})
missing_value_df.sort_values(by=['percent_missing'], ascending=False)
```

```python
# Number of set_num that is NaN

merged[merged['set_num'].isnull()].shape
```

```python
# Check if is_licensed is True
licensed = merged[merged['is_licensed']]

# Only drop the na in set_num columns
licensed = licensed.dropna(subset=['set_num'])
licensed.head()
```

```python
# Filter 'Star War'

star_wars = licensed[licensed['parent_theme']=='Star Wars']
star_wars.head()
```

```python
# Count of licenses and count of star wars to get the percentage 

the_force = float(star_wars.shape[0]/licensed.shape[0])*100

print(the_force)
```

### 2. Which year was Star Wars not the most popular licensed set(in terms of number of sets released that year)

```python
licensed['count']=1

# Count (using aggregate) to get the number of years that's licensed.
summed_df = licensed.groupby(['year','parent_theme']).sum().reset_index()

# Sort by descending of the same years and drop any duplicate years
max_df = summed_df.sort_values('count', ascending=False).drop_duplicates(['year'])

max_df.head()
```

```python
max_df.sort_values('year', inplace=True)
max_df
```

```python
# 2017 is Star Wars is not the most popular set
```

### What is the number of sets by year?

```python
# ~ means NOT, following code is filter out sets that are not null.

new_df = merged[~merged['set_num'].isnull()]
new_df.head()
```

```python
new_df['count']=1
new_df.groupby(['year']).sum().reset_index()[['year','count']]
```

```python
sets_per_year = new_df.groupby(['year']).sum().reset_index()[['year','count']]
sets_per_year.sort_values('count', ascending=False)
```

```python
for index, row in sets_per_year.iterrows():
  print(row['year'], row['count'])
```
