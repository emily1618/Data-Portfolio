```python
 # Import libraries

import pandas as pd
import seaborn as sns
import numpy as np
    
import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure
    
%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8) #ADjust the confi of the plots we will create.

# Read the data
df = pd.read_csv(r'C:\Users\xteni\Desktop\School\Projects\Python\movies.csv')
```

```python
# Look at the data

df.head()
```

```python
# Find missing data


# Check if col has null
df.isnull().sum()
for each_col in df.columns:
    percent_missing = np.mean(df[each_col].isnull()) 
    print('{} - {}%'.format(each_col, round(percent_missing*100)))
```

```python
# Data types for columns

df.dtypes
```

```python
# Eliminate the .0 into a number without .0 at the end

df = df.fillna(0)
df['budget'] = df['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64')
df['votes'] = df['votes'].astype('int64')

df.dtypes
```

```python
# Take release date year value into a new column for the year of movie - old version of the data
# df['year_released'] = df['released'].astype(str).str[0:4]
```

```python
# Sort by gross value
df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
# Check and drop duplicates 

df.drop_duplicates()
```

```python
# Check correlation to gross revenue

# 1. Budget is highly correlated to revenue, the more money the company spent, the more revenue they will bring in. 
# Use a scatterplot budget vs gross

plt.scatter(x=df['budget'], y=df['gross'])
plt.title('Budget VS Gross')
plt.xlabel('Budget for Movie')
plt.ylabel('Gross Earning')

plt.show


```

```python
# Using regression plot seaborn to check correlation

sns.regplot(x='budget', y='gross', data=df, scatter_kws={"color": "red"}, line_kws={"color":"green"})
```

```python
# Look at data correlation, default is pearson

df.corr(method='pearson')
```

```python
# Visualize the correlation matrix
# Higher correlation, the color is light beige. Lesser correlation, the color is black

correlation_matrix = df.corr(method='pearson')

sns.heatmap(correlation_matrix, annot=True) 

plt.title('Corelation Matrix Heat Map')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')

plt.show()
```

```python
# Look at company and convert to a number because company is a word
# Use cat.code to numerize the column name
# For each column that is a object, it will change from object type to a numerized assignment

df_numerized = df

for each_col_name in df_numerized.columns:
    if(df_numerized[each_col_name].dtype == 'object'):
        df_numerized[each_col_name] = df_numerized[each_col_name].astype('category')
        df_numerized[each_col_name] = df_numerized[each_col_name].cat.codes 
        
df_numerized.head()


# From this heatmap, you can see gross and budget is highly correlated, gross and votes as well.
correlation_matrix = df_numerized.corr(method='pearson')

sns.heatmap(correlation_matrix, annot=True) 

plt.title('Corelation Matrix Heat Map')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')

plt.show()
```

```python
#Find highest correlation quickly using unstacking

correlation_mat = df_numerized.corr()
corr_pairs = correlation_mat.unstack()
corr_pairs

sorted_pairs = corr_pairs.sort_values()
sorted_pairs

#High correlation pairs (>0.50)
high_corr = sorted_pairs[sorted_pairs>0.5]
high_corr

#Conclusion: votes and budget have highest correlation to gross earning
```
